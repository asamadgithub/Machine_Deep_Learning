```python
fig, ax = plt.subplots(figsize = (10,6))

x = np.linspace(0, 1, 1002)[1:-1]


ax.plot(x ,beta(1, 1).pdf(x), linestyle = "solid", color = "black", alpha = 1.00,  label = f"a = 1.5, b = 1.5" )
ax.plot(x ,beta(1.1, 1.1).pdf(x), linestyle = "dashed", color = "black", alpha = 0.60,  label = f"a = 1.1, b = 1.1" )


ax.plot(x ,beta(1.5, 1.5).pdf(x), linestyle = "dashdot", color = "maroon", alpha = 0.20,  label = f"a = 1.5, b = 1.5" )
ax.plot(x ,beta(3, 3).pdf(x)    , linestyle = "dashdot", color = "maroon", alpha = 0.40,  label = f"a = 1.5, b = 1.5" )
ax.plot(x ,beta(4, 4).pdf(x)    , linestyle = "dashdot", color = "maroon", alpha = 0.60,  label = f"a = 1.5, b = 1.5" )
ax.plot(x ,beta(7, 7).pdf(x)    , linestyle = "dashdot", color = "maroon", alpha = 0.80,  label = f"a = 1.5, b = 1.5" )
ax.plot(x ,beta(10, 10).pdf(x)  , linestyle = "dashdot", color = "maroon", alpha = 1.00,  label = f"a = 1.5, b = 1.5" )


ax.set_title("Probability density function for 'Beta' distribution", size = 15, color  = "green")
ax.set_xlabel("theta", size = 12)
ax.set_ylabel("probability", size = 12)
# plt.xlabel('$theta$')
ax.set_ylim(0,4)

ax.legend(loc=0)
```
